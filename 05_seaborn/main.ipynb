## Lineplot üìà

- `lineplot`: Line charts are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group.

````python
df = pd.DataFrame({
    'a':[10,20,30,40,35],
    'b':[4,13,40,25,15],
    'c':[25,36,10,29,45],
    'd':['Deependu','Rahul','Raj','Rohit','Rajesh']
})

# Set the width and height of the figure
plt.figure(figsize=(14,6)) # 14 inches wide and 6 inches tall

# Add title
plt.title("Title of the plot")

plt.xlabel("X label")
plt.ylabel("Y label")

sns.lineplot(data=df, markers=True)
````

## Scatter Plot üìâ

- `scatterplot`: Scatter plots show the relationship between two continuous variables; if color-coded, we can also show the relationship with a third categorical variable.

- `regplot (regression plot)`: Regression plots are used to fit a linear regression model and show the strength of the relationship between the variables.

- `lmplot (linear model plot)`: lmplot combines regplot() and FacetGrid.

- `swarmplot`: Categorical scatter plots show the relationship between a continuous variable and a categorical variable.

‚ö†Ô∏è Note:
    - To color-code the scatter plots, use the argument **`hue`** with the name of the categorical variable.


````python
# Path of the file to read
insurance_filepath = "insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)

insurance_data.head()
````

````python
# scatter plot
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
````

````python
# color coding the scatter plot based on the 'smoker' (categorical variable) column
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
````

````python
# regplot
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'], scatter_kws={"color": "green"}, line_kws={"color": "red"})
````

````python
# lmplot
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
````

````python
# swarmplot
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
plt.grid()
````

## Plotting Distributions üìà

### Histogram üìä

````python
iris_dataset = './Iris.csv'
iris_data = pd.read_csv(iris_dataset)
iris_data.head()
````

````python
# Histogram
sns.histplot(data=iris_data, x='PetalLengthCm', hue='Species')
````

---

### KDE (Kernel Density Estimate) Plot üìà

- a smoothed histogram.

````python
sns.kdeplot(data=iris_data, x='PetalLengthCm', hue='Species', fill=True)
````

---

### 2D-KDE Plot üìà

- We're not restricted to a single column when creating a KDE plot. We can create a two-dimensional (2D) KDE plot with the sns.jointplot command.

- `the curve at the top of the figure is a KDE plot for the data on the x-axis` (in this case, iris_data['Petal Length (cm)']), and

- `the curve on the right of the figure is a KDE plot for the data on the y-axis` (in this case, iris_data['Sepal Width (cm)']).



````python
# 2D KDE plot
sns.jointplot(x=iris_data['PetalLengthCm'], y=iris_data['SepalWidthCm'], kind="kde")
````

## Styling seaborn Plots

- We can quickly change the style of the figure to a different theme with only a single line of code.

```python
# Change the style of the figure to the "dark" theme
sns.set_style("dark")
```

- Seaborn has five different themes: `(1)"darkgrid", (2)"whitegrid", (3)"dark", (4)"white", and (5)"ticks"`, and you need only use a command similar to the one in the code cell above (with the chosen theme filled in) to change it.

````python
df = pd.DataFrame({
    'a':[10,20,30,40,35],
    'b':[4,13,40,25,15],
    'c':[25,36,10,29,45],
    'd':['Deependu','Rahul','Raj','Rohit','Rajesh']
})

# Set the width and height of the figure
plt.figure(figsize=(14,6)) #  sets the size of the figure to 14 inches (in width) by 6 inches (in height)
sns.set_style("darkgrid") # darkgrid, whitegrid, dark, white, ticks

sns.lineplot(data=df, label="My Data")
````
