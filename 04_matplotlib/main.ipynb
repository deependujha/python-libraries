## PyPlot ðŸ“Š

- Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

> ## Create a simple plot

    - We pass two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis.
    - If we pass only one array, then that array is assumed to be the y-axis values, and the x-axis will be the indices of the elements in the array.


````python
## passing two arrays

xPoints = np.array([0, 2,3,7,9])
yPoints = np.array([10,15,7,20,30])

plt.plot(xPoints, yPoints)
plt.show() # show the plot
````

````python
## passing one array

yPoints = np.array([10,15,7,20,2])

plt.plot(yPoints)
plt.show() # show the plot
````

## Marker size & color:
```
- 'ms' for marker size
- 'mfc' for marker face color
- 'mec' for marker edge color
```

````python
yPoints = np.array([3, 8, 1, 10])

plt.plot(yPoints, marker = 'o', ms = 10, mec = 'r', mfc = 'g') # we can also use hex color code instead of 'r' and 'g'
plt.show()
````

---

## Line style & color:

- 'linestyle' or 'ls' for line style.  (dashed, dotted, dashdot, solid, None)

```python
# all four commands do the same
plt.plot(yPoints, linestyle = 'dotted')
plt.plot(yPoints, ls = 'dotted')
plt.plot(yPoints, linestyle = ':')
plt.plot(yPoints, ls = ':')
```

- 'color' or 'c' for line color.

```python
# all four commands do the same
plt.plot(yPoints, color = 'r')
plt.plot(yPoints, c = 'r')
plt.plot(yPoints, color = '#4CAF50')
plt.plot(yPoints, c = '#4CAF50')
```

- 'linewidth' or 'lw' for line width.

```python
# both commands do the same
plt.plot(yPoints, linewidth = 20)
plt.plot(yPoints, lw = 20)
```

````python
yPoints = np.array([3, 8, 1, 10])

plt.plot(yPoints, marker = '*', ms = 20, mec='r', mfc='g', c='c', ls=':', lw=2)
````

## Multiple plots ðŸ“ˆðŸ“‰

- You can plot as many plots as you want by simply adding more plt.plot() functions:

- When you are done with the plot, remember to call plt.show() to make it visible.


````python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])

x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, marker='o', ms=10, mec='r', mfc='g', c='c', ls=':', lw=2)
plt.plot(x2, y2, marker='*', ms=20, mec='r', mfc='b', c='k', ls=':', lw=2)

plt.show()
````

## Labels & Title ðŸ“

- To add title, we use the title() function in pyplot module.
- You can add labels to the axes with the xlabel() and ylabel() functions.

````python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
````

### Positioning the title

- You can use the loc parameter in title() function to position the title.
- Legal values are: 'left', 'right', and 'center'. Default value is 'center'.


````python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data", loc='left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
````

### font properties of title and labels ðŸ“

- You can use the fontdict parameter in title() and xlabel() functions to set font properties of title and labels.


````python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
````

## Subplots (Displaying multiple subplots in one plot) ðŸ“ŠðŸ“Š

- With the subplots() function you can draw multiple plots in one figure:

- The subplots() function takes three arguments that describes the layout of the figure. The layout is organized in rows and columns, which are represented by the first and second argument. The third argument represents the index of the current plot.

```python
plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
```

- We can add a title to each subplot with the title() function:

- Also, we can add a super title to the entire figure with the suptitle() function:

````python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.title('first plot')
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.title('second plot')
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.title('third plot')
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.title('fourth plot')
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.title('fifth plot')
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.title('sixth plot')
plt.plot(x,y)

plt.suptitle("My Subplot Example")
plt.show()
````

## Scatter plot ðŸ“Š

- A scatter plot is a diagram where each value in the data set is represented by a dot.

- To specify the **MARKER**, you can use the keyword argument **marker** or the shorter version **'m'**.

- You can also choose the **COLOR** of the dots. To do so, use the keyword argument **c** or the shorter version **'c'**.

- You can also choose the **SIZE** of the dots. To do so, use the keyword argument **s** or the shorter version **'s'**.

- You can also choose the **ALPHA** value, which makes the color more or less transparent. To do so, use the keyword argument **alpha** or the shorter version **'alpha'**.

````python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5, c='hotpink') # size can be a list of size of each point or a single value. Same for alpha and c.

plt.show()
````

### Colormap in scatter plotðŸŒˆ

- A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

- You can use the keyword argument **cmap** or **'cmap'** to add a colormap.

- When using a colormap, you need to specify a value from the colormap for each dot in the plot.

- ```Some colormaps are: 'viridis', 'plasma', 'inferno', 'magma', 'cividis', 'rainbow', 'jet', 'turbo'```

````python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis') # cmap is used to change the color scheme

plt.colorbar() # to show the color bar

plt.show()
````

````python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
````

## Pie chart ðŸ“Š

- A pie chart is a circular graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices.

- The size of each wedge is determined by comparing the value with all the other values, by using this formula:

- The value divided by the sum of all values: x/sum(x)

- ```Explode:``` To emphasize one or more slices, use the explode parameter.

````python
y = np.array([35, 25, 25, 15])
myLabels = ["Apples", "Bananas", "Cherries", "Dates"]
myColors = ["black", "hotpink", "b", "#4CAF50"]

myExplode = [0.2, 0, 0, 0]

plt.pie(y, labels = myLabels, colors = myColors, shadow=True, explode=myExplode)

plt.legend(title="Fruits") # to show the legend
plt.show() 
````
